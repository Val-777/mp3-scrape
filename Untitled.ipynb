```python
from bs4 import BeautifulSoup
import requests
import re

downpage = 'http://tut-audio.su/mp3-album-luz-y-sombra-download-5420720.html'
page = requests.get(downpage)
soup = BeautifulSoup(page.content, 'html.parser')

url = 'http://tut-audio.su'

# m = re.search('(?<=(mp3-album-))[-a-z]+(?=(-download))', downpage)
# album = m.group(0).replace('-',' ').title()
album = soup.find(id="titlealb").get_text()[:-14]
year = soup.find(id="dopinfoalb").find('p').find('b').get_text()
image = url + soup.find(id="imagesalb").get('src')
```

```python
t = soup.find_all("div", "player")[0]
_, _, offset, _ = t['data-mp3url'].split('/')
offset = int(offset)
for link in soup.find_all("div", "player"):
    track = {}
    track['interpret'], track['trackname'] = link['data-title'].split(' — ')
    _, _, number, filename = link['data-mp3url'].split('/')
    number = str(int(number)-offset+1)
    if len(number) == 1:
        number = '0' + number
    track['number'] = number
    track['album'] = album
    track['filename'] = '{} {}.mp3'.format(number, track['trackname'])
    track['']
    print(track)
```
